```python
"""2. Crea una funcion que se creciente, que reciba una matriZ que verifique si' sus lineas
se encuentran en forMa  creciente, devolver un si,
 si es creciente y devolver un no si no lo es"""

def matriz_creciente(matriz):
    for fila in matriz:
        # Si la fila tiene menos de dos elementos, se considera creciente
        if len(fila) < 2:
            continue
        # Iterar sobre la fila para verificar el orden
        for i in range(len(fila) - 1):
            # Si el elemento actual es mayor que el siguiente, no es creciente
            if fila[i] > fila[i+1]:
                return "No"
    return "Si"

# Ejemplo de uso
matriz_ejemplo = [[1, 2, 3],
                  [4, 5, 6],
                  [7, 8, 9]]
resultado = matriz_creciente(matriz_ejemplo)
print(f"La matriz es creciente: {resultado}")  # Salida: : Si

matriz_no_creciente = [[1, 3, 2],
                       [4, 5, 6],
                       [7, 8, 9]]
resultado = matriz_creciente(matriz_no_creciente)
print(f"La matriz es creciente: {resultado}")  # Salida:: No

matriz_vacia = []
resultado = matriz_creciente(matriz_vacia)
print(f"La matriz es creciente: {resultado}")  # Salida:  Si
```

```python
"""Ejercicio 3._ CREE UNA FUNCION CANT_VECES(matriz) la cual calcule la cantidad de veces que aparece en la matriz , se imprime en pantalla solo las veces pares ejm
  listar cant de veces que aparece:
  3 aparece 6 veces"""

def cant_veces(matriz):
    """
    Cuenta la cantidad de veces que aparece cada nmero en una matriz
    y luego imprime los resultados pares.
    """
    contadores = {}
    for fila in matriz:
        for elemento in fila:
            if elemento in contadores:
                contadores[elemento] += 1
            else:
                contadores[elemento] = 1

    for numero, cantidad in contadores.items():
        if cantidad % 2 == 0:
            print(f"{numero} aparece {cantidad} veces")

# Ejemplo de uso:
matriz = [[1, 2, 3],
          [4, 3, 5],
          [2, 3, 6],
          [1, 2, 3]]
cant_veces(matriz)
```

```python
"""ejercicio 4
Cree la funcion LISTAR (matriz) la cual imprima los valores de
la matriz (sin repetirse),
# Ejemplo de uso:
matriz = [[1, 2, 3],
          [4, 3, 5],
          [2, 3, 6],
          [1, 2, 3]]
listar valores de M : 1,2,3,4"""
def LISTAR(matriz):
  """
  Recibe una matriz y la recorre, imprimiendo cada valor sin repeticiones.
  """
  valores_vistos = set()  # Usamos un conjunto para evitar repeticiones

  for fila in matriz:
    for valor in fila:
      if valor not in valores_vistos:
        print(valor, end=",") # Imprimir sin salto de lnea y con coma
        valores_vistos.add(valor)  # Agregamos el valor al conjunto

# Ejemplo de uso:
matriz = [[1, 2, 3],
          [4, 3, 5],
          [2, 3, 6],
          [1, 2, 3]]

print( end="")
LISTAR(matriz)
print("") # Imprimir una nueva lnea para finalizar la salida
```

```python
"""ejercicio 5 Cree la funcion ESPIRAL(N) la cual construya una matriz con N
 filas x N columnas cuyas entradas sean los numeros 1,2,3 (N*N) acomodados en
  espiral, comenzando con el numero 1 en el centro de la matriz use la funcion
   IMPRIMIR(matriz) para la impresion de la matriz
Ejemplo para un N=7"""
def ESPIRAL(N):
    # Crear matriz N x N llena de ceros
    matriz = [[0] * N for _ in range(N)]

    # Coordenadas del centro
    x, y = N // 2, N // 2

    # Valor inicial
    valor = 1
    matriz[x][y] = valor
    valor += 1

    # Direcciones: derecha, abajo, izquierda, arriba
    direcciones = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    paso = 1
    while valor <= N * N:
        for d in range(4):
            dx, dy = direcciones[d]
            for _ in range(paso + (d // 2 == 1)):
                x += dx
                y += dy
                if 0 <= x < N and 0 <= y < N:
                    matriz[x][y] = valor
                    valor += 1
        paso += 2

    return matriz

def IMPRIMIR(matriz):
    for fila in matriz:
        print(" ".join(f"{num:02}" for num in fila))

# Ejemplo de uso
matriz = ESPIRAL(7)
IMPRIMIR(matriz)
```
